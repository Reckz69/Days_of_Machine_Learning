```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
web = requests.get('https://books.toscrape.com/catalogue/category/books/psychology_26/index.html').text
```

```python
soup = BeautifulSoup(web, 'lxml')
```

```python
print(soup.prettify())
```

```python

books = soup.find_all('article', class_= 'product_pod')
```

```python
books
```

```python

data = []

for book in books:
    # Title
    title = book.h3.a["title"]

    # Price
    price = book.find("p", class_="price_color").text.strip()

    # Rating (stored in class)
    rating = book.find("p", class_="star-rating")["class"][1]

    # Availability
    availability = book.find("p", class_="instock availability").text.strip()

    data.append([title, price, rating, availability])

df = pd.DataFrame(data, columns=["Title", "Price", "Rating", "Availability"])
```
